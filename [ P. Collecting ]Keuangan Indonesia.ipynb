# Data BPS Suku Bunga Kredit Bank

#### ONLY SPECIFIC YEAR

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# Generate the number based on the year logic
# year = int(input("Enter the year (2002-2023): "))
number = 5

if number >= 1 and number <= 22:
    # Calculate the corresponding year
    year = 2024 - number

    # Make a GET request to the website
    url = f"https://www.bps.go.id/indicator/13/383/{number}/suku-bunga-kredit-rupiah-menurut-kelompok-bank.html"
    response = requests.get(url)

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the div containing the table
    div_table = soup.find("div", class_="table-responsive")

    if div_table:
        # Find all rows in the table
        rows = div_table.find_all("tr")

        # Extract the data for the specified year
        data = []
        if rows:
            for row in rows:
                # Extract the required columns from the row
                columns = row.find_all("td")
                data_series = [column.text.strip() for column in columns]
                data.append(data_series)

            # Create a dataframe from the data
            df = pd.DataFrame(data)

            # Transpose the dataframe
            df_transposed = df.transpose()

            # Set the first row as the column headers
            new_header = df_transposed.iloc[0]
            df_transposed = df_transposed[1:]
            df_transposed.columns = new_header

            # Drop row 13 and columns with None values
#             df_transposed = df_transposed.drop([13])
            df_transposed = df_transposed.dropna(axis=1)

            # Add a new column filled with the corresponding years
            df_transposed["Year"] = year

            # Print the transposed dataframe
            print(df_transposed)
        else:
            print("Data not available for the specified year.")
    else:
        print("Table not found on the webpage.")
else:
    print("Invalid year. Year must be between 2002 and 2023.")
```

#### YEAR 2023 TILL 2002

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# Create an empty dataframe to store the scraped data
df = pd.DataFrame()

# Loop through the numbers from 1 to 10
for number in range(1, 6):
    if number >= 1 and number <= 22:
        # Calculate the corresponding year
        year = 2024 - number

        # Make a GET request to the website
        url = f"https://www.bps.go.id/indicator/13/383/{number}/suku-bunga-kredit-rupiah-menurut-kelompok-bank.html"
        response = requests.get(url)

        # Create a BeautifulSoup object to parse the HTML content
        soup = BeautifulSoup(response.content, "html.parser")

        # Find the div containing the table
        div_table = soup.find("div", class_="table-responsive")

        if div_table:
            # Find all rows in the table
            rows = div_table.find_all("tr")

            # Extract the data for the specified year
            data = []
            if rows:
                for row in rows:
                    # Extract the required columns from the row
                    columns = row.find_all("td")
                    data_series = [column.text.strip() for column in columns]
                    data.append(data_series)

                # Create a dataframe from the data
                df_new = pd.DataFrame(data)

                # Transpose the dataframe
                df_new_transposed = df_new.transpose()

                # Set the first row as the column headers
                new_header = df_new_transposed.iloc[0]
                df_new_transposed = df_new_transposed[1:]
                df_new_transposed.columns = new_header

                # Drop row 13 and columns with None values
#                 df_new_transposed = df_new_transposed.drop([13])
                df_new_transposed = df_new_transposed.dropna(axis=1)

                # Add a new column filled with the corresponding year
                df_new_transposed["Year"] = year

                # Append the current dataframe to the main dataframe
                df = df.append(df_new_transposed, ignore_index=True)
            else:
                print(f"Data not available for the specified year {year}.")
        else:
            print(f"Table not found on the webpage for the specified year {year}.")
    else:
        print("Invalid year. Year must be between 2002 and 2023.")

# Print the resulting dataframe
print(df)
```

```python
df = df.replace('-',np.nan)
df_IR = df.dropna()
df_IR
```

```python
df_IR['Year'] = pd.to_datetime(df_IR['Year'], format='%Y')
df_IR['Month'] = df_IR.groupby(df_IR['Year'].dt.year).cumcount().apply(lambda x: (pd.Timestamp(year=1900, month=x % 12 + 1, day=1)).strftime('%B'))
df_IR
```
